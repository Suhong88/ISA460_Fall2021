The sparksession instance is the way Spark executes user-defined manipulation across the cluster

```
# start a spark session
spark
```

```
# perform a simple task of creating a range of numbers and covert it to a dataframe.
# Dataframe concept is similar to Pandas in python. You can convert dataframe to pandas, and vise sersa.
myRange= spark.range(1000).toDF("number")
myRange.show()

pd=myRange.toPandas() #covert dataframe to Pandas
pd # dispplay pd

df = spark.createDataFrame(pd) #covert pandas to dataframe
```
